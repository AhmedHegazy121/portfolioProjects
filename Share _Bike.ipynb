Make columns consistent and merge them into a single dataframe

```python
bike_list = [share_Bike_1,share_Bike_2,share_Bike_3,share_Bike_4,share_Bike_5,share_Bike_6,share_Bike_7,share_Bike_8,share_Bike_9,share_Bike_10,share_Bike_11,share_Bike_12]
all_bike =pd.concat(bike_list)
all_bike.info()
```

```python
all_bike.head()
```

Connvert  start time and end time  form objest into Date

```python
all_bike['started_at'] = pd.to_datetime(all_bike['started_at'])
all_bike['ended_at'] = pd.to_datetime(all_bike['ended_at'])
all_bike.info()
```

```python
all_bike.rename(columns={'rideable_type':'ride_type','started_at':'start_time','ended_at':'end_time','member_casual':'member_type'},inplace= True )
all_bike.head()
```

Create Colummns to identfy  time  , months , year and day of the weeks

```python
#Time 
all_bike['time']= all_bike['start_time'].dt.strftime("%H:%M")
#month
all_bike['month']=all_bike['start_time'].dt.month_name()
#year
all_bike['year']=all_bike['start_time'].dt.year
#Day of the week 
all_bike['day_of_week']=all_bike['start_time'].dt.day_name()

all_bike.head()
```

```python
#combining 'month' & 'year' column
all_bike['month_year']=pd.to_datetime(all_bike['month'].astype(str) + all_bike['year'].astype(str),format ='%B%Y').dt.strftime('%b-%Y')
```

```python
all_bike
```

Calculate the length of each ride

```python
all_bike['length_time'] = (all_bike['end_time']-all_bike['start_time']).dt.total_seconds().div(60).astype(float)
all_bike.head()
```

Dropping rows containing Trip Duration in Negative


```python
# check if there  are Negative values
print(len(
    all_bike[all_bike['length_time']<0]
    ))
```

```python
df_index=all_bike[all_bike['length_time']<0].index
bike =all_bike.drop(df_index)

bike.info()
```

Dropping rows containing Tests in start_station_name



```python
print(len(
    bike[bike['start_station_name'].str.contains('Test'or 'test' or 'TEST' )==True]
))
```

```python
df_index =bike[bike['start_station_name'].str.contains('Test'or 'test' or 'TEST' )==True].index
bike = bike.drop(df_index)
```

```python
bike.head()
```

Drop the Columns that we don't need  (year , month , end_ing  and start_lat )

```python
# drop some columns
bike.drop(['year', 'month', 'end_lng', 'start_lat'], axis=1, inplace=True)
bike.head()
```

Checking for missing data

```python
print(f'Tota of missing value : {bike['ride_type'].isna().sum()}')
print(f'Tota of missing value : {bike['ride_type'].isna().sum()}')
print(f'Tota of missing value : {bike['start_time'].isna().sum()}')
print(f'Tota of missing value : {bike['end_time'].isna().sum()}')
print(f'Tota of missing value : {bike['member_type'].isna().sum()}')

```
